#SOAL 1

```python
class Customer:
    def __init__(self, ticket_id):
        self.ticket_id = ticket_id

class Queue:
    def __init__(self, capacity=10):
        self.capacity = capacity
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, ticket_id):
        if self.size() < self.capacity:
            new_customer = Customer(ticket_id)
            self.queue.append(new_customer)
            print(f"Pelanggan {ticket_id} ditambahkan ke antrean.")
        else:
            print("Antrean sudah penuh. Mohon menunggu hingga slot tersedia.")

    def dequeue(self):
        if not self.is_empty():
            served_customer = self.queue.pop(0)
            print(f"Pelanggan {served_customer.ticket_id} telah dilayani.")
            return served_customer
        else:
            print("Antrean kosong. Tidak ada pelanggan yang harus dilayani.")
            return None

    def front(self):
        if not self.is_empty():
            return self.queue[0].ticket_id
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1].ticket_id
        else:
            return None

    def serve_vip(self, ticket_id):
        if '7' in ticket_id:
            print(f"Pelanggan VIP {ticket_id} akan dilayani segera.")
        else:
            print(f"Pelanggan {ticket_id} bukan VIP.")

# Contoh penggunaan
if __name__ == "__main__":
    queue = Queue()

    # Menambahkan pelanggan
    queue.enqueue("NA59")  # Pelanggan Anda
    queue.enqueue("AB01")  # Pelanggan biasa
    queue.enqueue("CD02")  # Pelanggan biasa
    queue.enqueue("EF03")  # Pelanggan biasa
    queue.enqueue("GH04")  # Pelanggan biasa
    queue.enqueue("IJ05")  # Pelanggan biasa
    queue.enqueue("MN07")  # Pelanggan VIP
    queue.enqueue("OP08")  # Pelanggan biasa
    queue.enqueue("QR09")  # Pelanggan biasa
    queue.enqueue("ST10")  # Pelanggan biasa
    queue.enqueue("UV11")  # Antrian penuh

    # Menampilkan pelanggan di depan dan belakang antrian
    print(f"Antrean Depan: {queue.front()}")
    print(f"Antrean Belakang: {queue.rear()}")

    # Melayani pelanggan
    queue.dequeue()
    queue.dequeue()

    # Menambahkan pelanggan lagi
    queue.enqueue("WX12")  # Pelanggan biasa

    # Melayani pelanggan VIP
    queue.serve_vip("MN07")  # Pelanggan VIP
```

#SOAL 2

```python
class Order:
    def __init__(self, order_id, num_pages):
        self.order_id = order_id
        self.num_pages = num_pages

class Queue:
    def __init__(self, capacity=10):
        self.capacity = capacity
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, order_id, num_pages):
        if self.size() < self.capacity:
            new_order = Order(order_id, num_pages)
            self.queue.append(new_order)
            print(f"Pesanan {order_id} dengan {num_pages} halaman telah ditambahkan ke antrian.")
        else:
            print("Antrian penuh. Silakan tunggu hingga ada slot yang tersedia.")

    def dequeue(self):
        if not self.is_empty():
            served_order = self.queue.pop(0)
            print(f"Pesanan {served_order.order_id} telah dilayani.")
            # Jika pesanan memiliki lebih dari 100 halaman, pisahkan menjadi dua tahap
            if served_order.num_pages > 100:
                remaining_pages = served_order.num_pages - 100
                print(f"Pesanan {served_order.order_id} memiliki {remaining_pages} halaman tersisa dan akan ditambahkan kembali ke antrian.")
                self.enqueue(served_order.order_id, remaining_pages)
            return served_order
        else:
            print("Antrian kosong. Tidak ada pesanan untuk dilayani.")
            return None

    def front(self):
        if not self.is_empty():
            return self.queue[0].order_id
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1].order_id
        else:
            return None

    def process_fast_track(self, order_id, num_pages):
        if num_pages < 10:
            print(f"Fast Track: Pesanan {order_id} dengan {num_pages} halaman sedang diproses segera.")
        else:
            print(f"Pesanan {order_id} dengan {num_pages} halaman tidak memenuhi syarat untuk Fast Track.")

# Contoh penggunaan
if __name__ == "__main__":
    queue = Queue()

    # Menambahkan pesanan fotokopi
    queue.enqueue("NAV59", 5)   # Pesanan Fast Track
    queue.process_fast_track("NAV59", 5)  # Memproses Fast Track

    queue.enqueue("NAV59", 150)  # Pesanan lebih dari 100 halaman
    queue.enqueue("NAV59", 50)    # Pesanan biasa
    queue.enqueue("NAV59", 8)     # Pesanan Fast Track
    queue.process_fast_track("NAV59", 8)  # Memproses Fast Track

    # Menampilkan pesanan di depan dan belakang antrian
    print(f"Pesanan di depan: {queue.front()}")
    print(f"Pesanan di belakang: {queue.rear()}")

    # Melayani pesanan
    queue.dequeue()  # Melayani pesanan pertama
    queue.dequeue()  # Melayani pesanan kedua

    # Menampilkan sisa pesanan
    print(f"Sisa pesanan dalam antrian: {queue.size()}")
```

```python
class Vehicle:
    def __init__(self, vehicle_number):
        self.vehicle_number = vehicle_number

class Queue:
    def __init__(self, capacity=15):
        self.capacity = capacity
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, vehicle_number):
        if self.size() < self.capacity:
            new_vehicle = Vehicle(vehicle_number)
            self.queue.append(new_vehicle)
            print(f"Kendaraan {vehicle_number} telah masuk ke antrian parkir.")
        else:
            print("Parkiran penuh. Silakan tunggu hingga ada kendaraan yang keluar.")

    def dequeue(self):
        if not self.is_empty():
            served_vehicle = self.queue.pop(0)
            print(f"Kendaraan {served_vehicle.vehicle_number} telah keluar dari parkiran.")
            return served_vehicle
        else:
            print("Parkiran kosong. Tidak ada kendaraan untuk dikeluarkan.")
            return None

    def front(self):
        if not self.is_empty():
            return self.queue[0].vehicle_number
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1].vehicle_number
        else:
            return None

    def park_priority(self, vehicle_number):
        # Misalkan kendaraan ramah lingkungan ditandai dengan awalan "EL" (Electric)
        if vehicle_number.startswith("EL"):
            print(f"Parkir Prioritas: Kendaraan {vehicle_number} mendapatkan prioritas masuk parkiran.")
            self.enqueue(vehicle_number)
        else:
            print(f"Kendaraan {vehicle_number} tidak mendapatkan prioritas parkir.")

# Contoh penggunaan
if __name__ == "__main__":
    queue = Queue()

    # Menambahkan kendaraan
    queue.park_priority("EL59")  # Kendaraan ramah lingkungan
    queue.park_priority("TA59")   # Kendaraan biasa
    queue.enqueue("TA01")         # Kendaraan biasa
    queue.enqueue("TA02")         # Kendaraan biasa
    queue.enqueue("TA03")         # Kendaraan biasa
    queue.enqueue("TA04")         # Kendaraan biasa
    queue.enqueue("TA05")         # Kendaraan biasa
    queue.enqueue("TA06")         # Kendaraan biasa
    queue.enqueue("TA07")         # Kendaraan biasa
    queue.enqueue("TA08")         # Kendaraan biasa
    queue.enqueue("TA09")         # Kendaraan biasa
    queue.enqueue("TA10")         # Kendaraan biasa
    queue.enqueue("TA11")         # Kendaraan biasa
    queue.enqueue("TA12")         # Kendaraan biasa
    queue.enqueue("TA13")         # Kendaraan biasa
    queue.enqueue("TA14")         # Kendaraan biasa
    queue.enqueue("TA15")         # Kendaraan biasa
    queue.enqueue("TA16")         # Parkiran penuh

    # Menampilkan kendaraan di depan dan belakang antrian
    print(f"Kendaraan di depan: {queue.front()}")
    print(f"Kendaraan di belakang: {queue.rear()}")

    # Mengeluarkan kendaraan
    queue.dequeue()  # Mengeluarkan kendaraan pertama
    queue.dequeue()  # Mengeluarkan kendaraan kedua

    # Menampilkan sisa kendaraan
    print(f"Sisa kendaraan dalam antrian: {queue.size()}")
```

```python
class Student:
    def __init__(self, student_id):
        self.student_id = student_id

class Queue:
    def __init__(self, capacity=20):
        self.capacity = capacity
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, student_id):
        if self.size() < self.capacity:
            new_student = Student(student_id)
            self.queue.append(new_student)
            print(f"Mahasiswa {student_id} telah ditambahkan ke antrian pengambilan sertifikat.")
        else:
            print("Antrian penuh. Silakan tunggu hingga ada slot yang tersedia.")

    def dequeue(self):
        if not self.is_empty():
            served_student = self.queue.pop(0)
            print(f"Mahasiswa {served_student.student_id} telah dilayani.")
            return served_student
        else:
            print("Antrian kosong. Tidak ada mahasiswa untuk dilayani.")
            return None

    def front(self):
        if not self.is_empty():
            return self.queue[0].student_id
        else:
            return None

    def rear(self):
        if not self.is_empty():
            return self.queue[-1].student_id
        else:
            return None

    def process_priority(self, student_id):
        # Misalkan mahasiswa berprestasi ditandai dengan awalan "PR" (Prestasi)
        if student_id.startswith("PR"):
            print(f"Mahasiswa {student_id} berprestasi, langsung mengambil sertifikat tanpa antri.")
        else:
            self.enqueue(student_id)

# Contoh penggunaan
if __name__ == "__main__":
    queue = Queue()

    # Menambahkan mahasiswa
    queue.process_priority("T59")   # Mahasiswa biasa
    queue.process_priority("PR59")   # Mahasiswa berprestasi
    queue.enqueue("T01")              # Mahasiswa biasa
    queue.enqueue("T02")              # Mahasiswa biasa
    queue.enqueue("T03")              # Mahasiswa biasa
    queue.enqueue("T04")              # Mahasiswa biasa
    queue.enqueue("T05")              # Mahasiswa biasa
    queue.enqueue("T06")              # Mahasiswa biasa
    queue.enqueue("T07")              # Mahasiswa biasa
    queue.enqueue("T08")              # Mahasiswa biasa
    queue.enqueue("T09")              # Mahasiswa biasa
    queue.enqueue("T10")              # Mahasiswa biasa
    queue.enqueue("T11")              # Mahasiswa biasa
    queue.enqueue("T12")              # Mahasiswa biasa
    queue.enqueue("T13")              # Mahasiswa biasa
    queue.enqueue("T14")              # Mahasiswa biasa
    queue.enqueue("T15")              # Mahasiswa biasa
    queue.enqueue("TS16")              # Mahasiswa biasa
    queue.enqueue("T17")              # Mahasiswa biasa
    queue.enqueue("T18")              # Mahasiswa biasa
    queue.enqueue("T19")              # Mahasiswa biasa
    queue.enqueue("T20")              # Mahasiswa biasa
    queue.enqueue("T21")              # Antrian penuh

    # Menampilkan mahasiswa di depan dan belakang antrian
    print(f"Mahasiswa di depan: {queue.front()}")
    print(f"Mahasiswa di belakang: {queue.rear()}")

    # Melayani mahasiswa
    queue.dequeue()  # Melayani mahasiswa pertama
    queue.dequeue()  # Melayani mahasiswa kedua

    # Menampilkan sisa mahasiswa
    print(f"Sisa mahasiswa dalam antrian: {queue.size()}")
```
